## Problem 1.  What <i>kind</i> of distribution does the number of Beneficiaries appear to follow?

```python
%%R -w 600 -h 400




NumberOfBeneficiaries = NumericData[,2]
#x <- rgamma(500, shape = 5, rate = 0.1)

# intuitively, it would make sense if this number followed some kind of arrival distribution

# The number of beneficiaries is very large in some cases (see the summary above),
# so let's try a plot of log10(number of beneficiaries)
# to see how many there are

hist( log10(NumberOfBeneficiaries), probability=TRUE, border=NA, col="deepskyblue", breaks=100 )

#Predict how good gamma distribution fit this 
    


parameters1 = fitdistr( log10(NumberOfBeneficiaries), "gamma" )
#print(parameters1)
    
curve( dgamma(x, shape=parameters1$estimate[1], rate=parameters1$estimate[2]),col="blue", add=TRUE)
#mtext( "good grief, the gamma distribution doesn't fit well either", line=-1,side=3, col="blue" )

parameters2 = fitdistr( log10(NumberOfBeneficiaries), "lognormal" )
curve( dlnorm(x, meanlog=parameters2$estimate[1], sdlog=parameters2$estimate[2]), col="red", add=TRUE )  # doesn't fit very well
#mtext( "that's annoying, the lognormal distribution doesn't fit very well", side=3, col="red" )    


parameters3 = fitdistr( log10(NumberOfBeneficiaries), "normal" )
#print(parameters3)
curve( dnorm(x, mean=parameters3$estimate[1],sd=parameters3$estimate[2]), col="green", add=TRUE )  # doesn't fit very well
#mtext( "that's annoying, the lognormal distribution doesn't fit very well", side=3, col="red" )    
```

```python
%%R
# The histogram above shows that most numbers of beneficiaries are below 10^(3.7) = 5,000.

# The R expression  (NumberOfBeneficiaries < 5000)  yields a boolean vector.
# It can be used as a subscript to select all entries of a vector satisfying this boolean:

NumberOfBeneficiariesBelow5K = NumberOfBeneficiaries[ (NumberOfBeneficiaries < 5000) ]

hist( NumberOfBeneficiariesBelow5K,  col="deepskyblue", probability= TRUE, border=NA, breaks=500  )

parameters3 = fitdistr( NumberOfBeneficiariesBelow5K, "exponential" )
#print(parameters3)
curve( dexp(x, rate=parameters3$estimate[1]), col="green", add=TRUE )  # doesn't fit very well
#mtext( "that's annoying, the lognormal distribution doesn't fit very well", side=3, col="red" ) 

parameters4 = fitdistr( NumberOfBeneficiariesBelow5K, "lognormal" )
#print(parameters3)
curve( dlnorm(x, meanlog=parameters4$estimate[1],sdlog=parameters4$estimate[2]), col="red", add=TRUE )  # doesn't fit very well
#mtext( "that's annoying, the lognormal distribution doesn't fit very well", side=3, col="red" )
```

## Problem 5.  What is the distribution of maximum charges made by the different kinds of Providers?

```python
%%R -w 1100

max_charge_by_provider_type = aggregate( total_submitted_chrg_amt  ~  provider_type,  data=MedicarePaymentData,  max  )

hist( log10( max_charge_by_provider_type$total_submitted_chrg_amt ), col="purple", probability=TRUE )

rank_ordering_by_max_charge = order( log10(max_charge_by_provider_type[,2]), decreasing=TRUE )

ranked_by_type = max_charge_by_provider_type[ rank_ordering_by_max_charge , ]

print(ranked_by_type)

ranks = 1:nrow(ranked_by_type)

plot(   log(ranks), log(ranked_by_type$total_submitted_chrg_amt), type="l", col="blue" )
points( log(ranks), log(ranked_by_type$total_submitted_chrg_amt), col="darkblue", pch=20 )

text( log(ranks), log(ranked_by_type$total_submitted_chrg_amt),
     ranked_by_type$provider_type, pos=2, cex=0.5, col="red" )
```
