## Problem 1.  What <i>kind</i> of distribution does the number of Beneficiaries appear to follow?

```python
%%R -w 600 -h 400

NumberOfBeneficiaries = NumericData[,2]

# intuitively, it would make sense if this number followed some kind of arrival distribution

# The number of beneficiaries is very large in some cases (see the summary above),
# so let's try a plot of log10(number of beneficiaries)
# to see how many there are

hist(log10(NumberOfBeneficiaries),  col="deepskyblue", breaks=50, probability=TRUE)

parameters1 = fitdistr( log10(NumberOfBeneficiaries), "lognormal" )
curve( dlnorm(x, meanlog=parameters1$estimate[1], sdlog=parameters1$estimate[2]), col="red", add=TRUE )
mtext( "The lognormal distribution doesn't fit very well", side=3, col="red" )

parameters2 = fitdistr( log10(NumberOfBeneficiaries), "gamma")
curve( dgamma(x, shape=parameters2$estimate[1], rate=parameters2$estimate[2]), col="blue", add=TRUE )  # doesn't fit very well
mtext( "The gamma distribution doesn't fit well either", line=-1, side=3, col="blue" )


parameters3 = fitdistr( log10(NumberOfBeneficiaries), "normal")
curve(dnorm(x, mean=parameters3$estimate[1], sd =parameters3$estimate[2]),  col="black", add=TRUE )
mtext( "The normal distribution doesn't fit well ", line=-2, side=3, col="black" )

parameters4 = fitdistr( log10(NumberOfBeneficiaries), "weibull" )
curve( dweibull(x,shape=parameters4$estimate[1], scale=parameters4$estimate[2]), col="purple", add=TRUE )
mtext( "The weibull distribution fit very well", line=-3, side=3, col="purple" )

#compared with lognormal ,gama, normal and weibull distribution, weibull fits best




```

```python
%%R
# The histogram above shows that most numbers of beneficiaries are below 10^(3.7) = 5,000.

# The R expression  (NumberOfBeneficiaries < 5000)  yields a boolean vector.
# It can be used as a subscript to select all entries of a vector satisfying this boolean:

NumberOfBeneficiariesBelow5K = NumberOfBeneficiaries[ (NumberOfBeneficiaries < 5000) ]

hist( NumberOfBeneficiariesBelow5K,  col="deepskyblue", breaks=100, border=NA, probability=TRUE )
parameters1 = fitdistr( NumberOfBeneficiariesBelow5K, "weibull" )
curve( dweibull(x, shape=parameters1$estimate[1], scale=parameters1$estimate[2]), col="red", add=TRUE )
mtext( "The weibull distributionit fits very well", side=3, col="red" )
```

## Problem 5.  Does Total Charge Amount by State follow a Power Law?

```python
%%R -w 800 -h 800

total_charge_by_state = aggregate( total_submitted_chrg_amt  ~  nppes_provider_state,  data=MedicarePaymentData,  sum  )
#~means agains, so sum the total_submitted_chrg_amt grouped by state.

#print(total_charge_by_state )

hist( log( total_charge_by_state[,2] ), col="seagreen", breaks=20 )

#print(total_charge_by_state[,2])

ranked_by_state = total_charge_by_state[ order( total_charge_by_state[,2], decreasing=TRUE ), ]
#print(ranked_by_state)

# keep only the top 50 states (ignore international data)
ranked_by_state = ranked_by_state[1:50,]

#print (ranked_by_state)
ranks = 1:nrow(ranked_by_state)
#print(ranks)

# ... is this relationship a power law?  Is the plot essentially log-log-linear?
# (Does it look like "the rich get richer" ?)
#print (ranked_by_state$total_submitted_chrg_amt)
#print(ranked_by_state$nppes_provider_state)

plot(   log(ranks), log(ranked_by_state$total_submitted_chrg_amt), type="l", col="blue" )  # line segments first
points( log(ranks), log(ranked_by_state$total_submitted_chrg_amt), col="darkblue", pch=20 )  # add dots
#ranked_by_state's subset: nppes_provider_state
text( log(ranks), log(ranked_by_state$total_submitted_chrg_amt),
     ranked_by_state$nppes_provider_state, pos=2, cex=0.75, col="red" )

abline(lm(log(ranked_by_state$total_submitted_chrg_amt)~log(ranks)),col="green")
mtext( "The log log linear relationship doesn't fit well",side=3, line=-1, col="green" )
dpareto=function(x, lamda, a) 
{
   a*x^(lamda)
}

#I try to plot a power law curve and to see whether it fits the plot of log(ranks) vs. log(ranked_by_state$total_submitted_chrg_amt)
#well.
x<-seq(0,5,0.001)

# I tried power law with power of 2 and 2.8; it fits much better than linear log log model
lines(x,dpareto(x, lamda = 2, a=-0.25)+24,col="black")
mtext( "The power law relationship with lamda of 2 and a of -0.25 doesn't fit very well", line=-2, side=3, col="black" )
lines(x,dpareto(x, lamda = 2.8, a=-0.1)+24,col="red")
mtext( "The power law relationship with lamda of 2.8 and a of -0.1 fits well ", line=-3, side=3, col="red" )






```

## Problem 5.  What is the distribution of maximum charges made by the different kinds of Providers?

```python
%%R -w 1100

max_charge_by_provider_type = aggregate( total_submitted_chrg_amt  ~  provider_type,  data=MedicarePaymentData,  max  )
#print (max_charge_by_provider_type)

hist( log10( max_charge_by_provider_type$total_submitted_chrg_amt ), col="purple", probability=TRUE )
#The distribution of maximum charges made against the different kinds of Providers fits logistic distribution well

parameter = fitdistr(log10(max_charge_by_provider_type$total_submitted_chrg_amt), "logistic" )
curve( dlogis(x,location=parameter$estimate[1], scale=parameter$estimate[2]), col="blue", lwd=2, add=TRUE )
mtext("Maximum charges made by the different kinds of Providers fits logistic distribution well", side=3, col="red")

rank_ordering_by_max_charge = order( log10(max_charge_by_provider_type[,2]), decreasing=TRUE )


ranked_by_type = max_charge_by_provider_type[ rank_ordering_by_max_charge , ]

#print(ranked_by_type)

ranks = 1:nrow(ranked_by_type)
#print(ranks)

plot(   log(ranks), log(ranked_by_type$total_submitted_chrg_amt), type="l", col="blue" )
points( log(ranks), log(ranked_by_type$total_submitted_chrg_amt), col="darkblue", pch=20 )

text( log(ranks), log(ranked_by_type$total_submitted_chrg_amt),
     ranked_by_type$provider_type, pos=2, cex=0.5, col="red" )


```

```python
"Maximum charges made by the different kinds of Providers fits logistic distribution well"
```
