## Please fix 'lratio' so this code (for exploring gender differences) works:

```python
%%R -w 900 -h 900

# Now that we have figured out which names we should focus on (at least to
# begin with), we need to calculate yearly summaries for each of those names.

selected <- subset(bnames, name %in% selected_names)
# In informations for the names that has boys_n> 10 & girls_n> 10 and boys + girls > 0.4

#print( nrow(selected) / nrow(bnames) )

# For each name in each year, figure out total boys and total girls

bysex <- ddply(selected, c("name", "year"), summarize,
  boys = sum(percent[sex == "boy"]),
  girls = sum(percent[sex == "girl"]),
  .progress = "text"
)
#print(bysex)
#print(bysex$boys)
#print(bysex$girls)

# It's useful to have a symmetric means of comparing the relative abundance
# of boys and girls - the log ratio is good for this.
#print(bysex)

######  PROBLEM:  lratio wants to be defined differently here ---
######  it can have infinite values, and these are set to "null" (NA)

######  PLEASE define lratio to be some other measure of
######   the relative abundance of boys and girls that is always defined.
minimumG<-1
minimumB<-1
for(i in 1:nrow(bysex)){
    if(bysex$girls[i]!=0){
        if(minimumG>bysex$girls[i]){
            minimumG<-bysex$girls[i]
        }
    }
    if(bysex$boys[i]!=0){
        if(minimumB > bysex$boys[i]){
            minimumB<-bysex$boys[i]
        }
    }
}
#print(minimumG)
#print(minimumB)



bysex$lratio <- log10((bysex$boys+minimumB)/( bysex$girls+minimumG))
bysex$lratio[!is.finite(bysex$lratio)] <- NA
#print(bysex)



#What I want to do is that I want to measure the relative abundance of boys and girls be just calculate their difference, boys-girls
#abs(boys-girls)/(boys+girls)
#print(bysex)

#print(bysex)

#print(bysex$boys[100])
#print(bysex$girls[100])
#print(bysex$lratio)
# Explore the distribution of lratio for each name.  This is why a smaller
# number of names helps - we can more easily see the name associated with the
# pattern and use our knowledge of names to determine whether or not it's
# plausible that the name is used for both sexes.

print(qplot(lratio, reorder(name,lratio), data = bysex))
print(qplot(abs(lratio), reorder(name,lratio), data = bysex))



```

```python
%%R -w 900 -h 900
#  Using R, figure out how to print the history of Lesley-like names over time for each gender.

#  That is, treating all these names as equivalent, plot the boy vs. girl percentages of
#  these name occurrences over time.

#  Initially the boy percentage is about 90%, but this changes over time.
#bnames$percentage <-bnames$

#print(qplot(year, data=subset(bnames, bnames$names in Lesley_like_names)
#print(LesleyBnames)

Lesley_like_names = c("Leslie", "Lesley", "Leslee", "Lesli", "Lesly")
targetNames <-subset(bnames, name %in% Lesley_like_names& year<=1997)
#print(subset(targetNames, year>1930 & year<1950))

targetNames <-ddply(targetNames,.(year, sex),  sump)
#print(head(targetNames))
#print(subset(targetNames, year>1930 & year<1950))
targetNamesB <-subset(targetNames, sex=="boy")
targetNamesG<-subset(targetNames, sex=="girl")
#print(head(targetNamesB))
#print(head(targetNamesG))
#print(head(targetNamesB))
#print(head(targetNamesG))
#print(nrow(targetNamesB))
#print(nrow(targetNamesG))
#print(head(targetNames))
#print(targetNamesB$percent[1])
#print(targetNamesG$percent[1])

#print(targetNamesB$percent[52])
#print(targetNamesG$percent[52])
#x<-targetNamesG$percent[52]/(targetNamesB$percent[52]+targetNamesG$percent[52])
#y<-targetNamesB$percent[52]/(targetNamesB$percent[52]+targetNamesG$percent[52])
#print(x)
#print(y)

for(i in 1:nrow(targetNamesB)){
    targetNamesB$boyPercent[i]<-targetNamesB$percent[i]/(targetNamesB$percent[i]+targetNamesG$percent[i])
    targetNamesG$girlPercent[i]<-targetNamesG$percent[i]/(targetNamesB$percent[i]+targetNamesG$percent[i])
}

#print(subset(targetNames, year>1930 & year<1950))
for(i in 1:nrow(targetNames)){
   if( targetNames$sex[i]=="boy"){
        targetNames$percent[i]<-targetNamesB$boyPercent[i/2+1]
    }else{
        #print("asdf asf asf ")
        targetNames$percent[i]<-targetNamesG$girlPercent[i/2]
    }
    
}
#print(targetNames)
print(qplot(year, percent, data=targetNames, col=sex, geom="line"))





```

# Optional Problem:  Ignoring gender, are there any names whose percentages always trend up, and which appear in at least 15 different years?

### (In other words, with an "up-trending" name, its percentage values never decrease over time.)

```python
%%R


#print(head(bnames))
names_ignore_sex<-ddply(bnames, c("name", "year"), sump, .progress="text")
freq_name_candidate<-ddply(names_ignore_sex, c("name"), summarise, count=sum(length(year)))
freq_names<-subset(freq_name_candidate, count>=15)

candidate<-subset(names_ignore_sex, name %in% freq_names$name)                  
checker<-function(data){
       result<-c()
       checkName<-data$name[1]
       checkPercent<-data$percent[1]
       count<-0
       for(i in 1: nrow(data)){
   
          if(checkName==data$name[i]){
             if(count==-1){
  
            }else if(data$percent[i]>checkPercent){
                count<-count+1
                checkPercent<-data$percent[i]
            }else{
                count<--1
            }
        }else{
            if(count>=15){
                result<-append(result, checkName)
            }
            count<-1
            checkName<-data$name[i]
            checkPercent<-data$percent[i]
         }
      }
      result
}
    

result<-checker(candidate)
print(result)
                           


```

# Optional Problem for Wizards:  get Hadley's Analysis for Baby Names by States to work

```python
%%R

# library(ggplot2)
# library(plyr)

# Try and find names that vary a lot between the states.

bnames <- read.csv("baby-names-by-state.csv", stringsAsFactors = F)
bnames <- subset(bnames, !is.na(number))
bnames$state <- factor(bnames$state)

# Add number of births and convert to percentage
births <- read.csv("births.csv")
bnames <- merge(bnames, births, by = c("state", "year", "sex"))
bnames$percent <- bnames$number / bnames$births

# Extract only names that have appeared in at least 25% of possible years
# and states
bnames$namesex <- paste(bnames$name, bnames$sex, sep = "-")
counts <- ddply(bnames, c("namesex"), summarize,
  n = length(namesex),
  number = sum(number))
counts <- counts[order(-counts$number), ]
counts <- subset(counts, n > 1250 * 0.25 & number > 1e5)

top <- subset(bnames, namesex %in% counts$namesex)
show_name <- function(name) {
  one <- top[top$namesex == name, ]
  qplot(year, percent, data = one, geom = "line", group = state)
}

# Look for names where there is a lot of variation in pattern between states

# Correlation approach
bystate <- cast(top, namesex + year ~ state, value = "percent")

cors <- dlply(bystate, "namesex", function(df)
  cor(as.matrix(df[, -(1:2)]), use = "pairwise.complete.obs"))

arrange(ldply(cors, min, na.rm = T), V1)

# Modelling approach - seems to do much better
patterns <- dlply(top, c("namesex"), function(df) {
  lm(percent ~ factor(year), data = df, weight = sqrt(births))
}, .progress = "text")

rsq <- function(mod) {
  summarize(summary(mod), rsq = r.squared, sigma = sigma)
}
qual <- arrange(merge(ldply(patterns, rsq), counts), -rsq)

sub <- c(as.character(qual$namesex[seq(1, nrow(qual), by = 5)]), "Juan-boy")

interesting <- subset(bnames, namesex %in% sub)

print(interesting)
    
# write.table(interesting, "interesting-names.csv", sep = ",", row = F)
                                                   
```
