## [ESL] Chapter 3: Exercise 5

Consider that the ridge regression problem can be rewritten as:

$$
\hat \beta = argmin_{\beta} \left( \sum_{i=1}^{N} \left( y_i - \beta_0 - \sum_{j=1}^{p}x_{ij} \beta_j \right)^2 + \lambda \sum_{j=1}^{p}{\beta_j}^2 \right)
$$

This is equivalent to:

$$
\hat \beta = argmin_{\beta} \left( \sum_{i=1}^{N} \left( y_i - \beta_0 - \sum_{j=1}^{p} \overline{x}_{j} \beta_j - \sum_{j=1}^{p}(x_{ij} - \overline{x}_j) \beta_j \right)^2 + \lambda \sum_{j=1}^{p}{\beta_j}^2 \right)
$$

By making the subsitution:

$$
\begin{align}
\beta_0^c &\mapsto \beta_0 + \sum_{j=1}^p \overline{x}_j \beta_j \\
\beta_j^c &\mapsto \beta_j, j = 1, 2, \dots, p
\end{align}
$$

Then the objective function can be rewritten as:

$$
\hat \beta^c = argmin_{\beta^c} \left( \sum_{i=1}^{N} \left( y_i - \beta^c_0 - \sum_{j=1}^{p}(x_{ij} - \overline{x}_j) \beta^c_j \right)^2 + \lambda \sum_{j=1}^{p}{\beta_j^c}^2 \right)
$$

Note that if $\hat \beta^c$ is the minimizer of our modified ridge regression, then $\hat \beta$
is the minimizer of the original ridge regression.

The modified objective function merely shifted intercept, all other coefficients remain untouched.

## [ISL] Chapter 4: Exercises 3

Because we are to compare the values of discriminant in different classes,
we don't need to care about the denominator in $p_k(x)$, as they are the same
among all classes. Therefore we take the log of numerators and get:

$$
\delta_k(x) = \log (\pi_k) + \log(\frac{1}{\sigma_k \sqrt{2\pi_k}}) - \frac{x^2 - 2x\mu_k + \mu_k^2}{2\sigma_k^2}
$$

Since $\sigma_k$ may vary across classes, term $x^2$ cannot be ignored.
Then it is obvious that Bayes classifier $\delta_k(x)$ is quadratic.

# The Supervised Learning paradigm in R

## Example:  Supervised learning on the Iris dataset

```python
%%R

data(iris)
print(str(iris))

n <- dim(iris)[1]
p <- dim(iris)[2]

print(table(iris$Species))

#------------------------------------------------------------------------
#   generate a training set, and create a model on the training set
#------------------------------------------------------------------------

iris_ids <- (1:n)

training_set <- sample(iris_ids, 75)     # Generate a random sample of size 75
print(training_set)

testing_set  <- iris_ids[-training_set]  # The set complement of training_set
print(testing_set)

print(table(iris$Species[training_set]))    # Tabulate species for the training set
print(table(iris$Species[testing_set]))     # Tabulate species for the testing set

#  plot the data as a parallel coordinates plot  (parcoord is in the MASS package)

parcoord(iris[,1:4],  col=c("red", "green3", "blue")[unclass(iris$Species)])
```

## fit an LDA model to the training set data

```python
%%R

#  lda is in the MASS package

ldaModel <- lda( Species ~ Petal.Length + Petal.Width + Sepal.Length + Sepal.Width,
                 data = iris, subset = training_set )

# Equivalent abbreviation:

ldaModel <- lda( Species ~ .,  data = iris, subset = training_set )


print(ldaModel)
```

## Evaluate how well the LDA model predicts classes for the remaining data (i.e., the testing_set)

```python
%%R

# Make class predictions on the test dataset using our LDA model
#  and then check the predicted class results against the true class values.

predictions <- predict( ldaModel, iris[ testing_set ,] )

print(predictions$class)

incorrect.prediction  <-  (predictions$class != iris$Species[testing_set])

print(testing_set[ incorrect.prediction ])          # ids  for which prediction was incorrect

print(iris[ testing_set[incorrect.prediction], ])   # data for which prediction was incorrect

confusion_matrix <- table( iris$Species[testing_set], predictions$class )
print(confusion_matrix)

m <- length(testing_set)
print(m)

accuracy <- (m - sum(incorrect.prediction))/m
print(accuracy)
```

## Never mix the training set with the testing set !!

```python
%%R

#  out of curiousity, let us see how well the model works on the training set.
#  (It is a terrible idea to do this in general)

predictions <- predict( ldaModel, iris[ training_set, ] )
print(predictions$class)

incorrect.prediction  <-  (predictions$class != iris$Species[training_set] )

print(iris_ids[ incorrect.prediction ]) # ids  for which prediction was incorrect
print(iris[ incorrect.prediction , ])   # data for which prediction was incorrect

accuracy <- (n - sum(incorrect.prediction))/n
print(accuracy)

confusion_matrix <- table( iris$Species[training_set], predictions$class )
print(confusion_matrix)
```

## Training another kind of model:  the SVM

```python
%%R

# Now train an SVM model, and evaluate its accuracy

library(e1071)

svmModel <- svm( Species ~ ., data=iris, subset = training_set )
print(summary(svmModel))

svm.predictions <- predict( svmModel, iris[ testing_set ,] )

svm.incorrect.prediction  <-  (svm.predictions != iris$Species[testing_set])
print(testing_set[svm.incorrect.prediction])

svm.confusion.matrix <- table( iris$Species[testing_set], svm.predictions )
print(svm.confusion.matrix)

svm.accuracy <- (m - length(training_set[svm.incorrect.prediction])) / m
print(svm.accuracy)
```

```python
%%R
# c
glm.pred<-rep("Down",nrow(Weekly))
glm.probs<-predict(glm.fit, type = "response")
glm.pred[glm.probs>0.5]= "Up"
print(table(glm.pred, Direction))
#Overall fraction of correct observations
print(mean(glm.pred==Direction))

# Sensitivity = P(glm.pred=Up | Direction = UP) = 557 / (557 + 48) = 92%
# Specificity = P(glm.pred=Down | Direction = Down) = 54 / (54 + 430) = 11%
# The probability of correctly prediction is only 56% for this logistic regression model.
# But since we never know the prior probabilities, we can't really think of the matrix in this manner.
# We should look at our predictions when they are Up and Down and compare them with what the data actually shows.
# When our model predicts that the market will go Down, the model is correct 52% of the time and when the model predicts that 
# the market is going "Up", it will be correct only 43.5% of the time. The mode is correct about its Down prediction more 
# often than about its Up prediction. However, the accuracy of its predictions are still not very good.
```

```python
%%R
# d
train<- Weekly$Year<=2008
train.X<-Weekly[train,1:8]
train.Y<-Weekly[train,9]
test.X<-Weekly[!train,1:8]
test.Y<-Weekly[!train,9]

glm.fit2<-glm(Direction~Lag2, data = Weekly[train,], family = binomial)
print(summary(glm.fit2))
glm.probs2<-predict(glm.fit2, Weekly[!train,], type = "response")
glm.pred2<-rep("Down",length(test.Y))
glm.pred2[glm.probs2>0.5]<-"Up"
print(table(glm.pred2, test.Y))
print(mean(glm.pred2==test.Y))
# The fraction of correct predictions has gone up to 62.5% by eliminating other predictors besides Lag2.
```

```python
%%R
# e
library(MASS)
lda.fit<-lda(Direction~Lag2, subset = train, family = binomial)
lda.fit
lda.pred<-predict(lda.fit, Weekly[!train,])
lda.class<-lda.pred$class
print(table(lda.class, test.Y))
# The lda gives the exact same results as logistic regression in part d.
```

```python
%%R
# f
qda.fit<-qda(Direction~Lag2, subset = train, family = binomial)
qda.fit
qda.pred<-predict(qda.fit, Weekly[!train,])
qda.class<-qda.pred$class
print(table(qda.class, test.Y))
# The qda always predicts that the stock market goes Up.
```
