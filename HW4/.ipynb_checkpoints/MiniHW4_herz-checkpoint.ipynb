# The Supervised Learning paradigm in R

## Example:  Supervised learning on the Iris dataset

```python
%%R

data(iris)
print(str(iris))

n <- dim(iris)[1]
p <- dim(iris)[2]

print(table(iris$Species))

#------------------------------------------------------------------------
#   generate a training set, and create a model on the training set
#------------------------------------------------------------------------

iris_ids <- (1:n)

training_set <- sample(iris_ids, 75)     # Generate a random sample of size 75
print(training_set)

testing_set  <- iris_ids[-training_set]  # The set complement of training_set
print(testing_set)

print(table(iris$Species[training_set]))    # Tabulate species for the training set
print(table(iris$Species[testing_set]))     # Tabulate species for the testing set

#  plot the data as a parallel coordinates plot  (parcoord is in the MASS package)

parcoord(iris[,1:4],  col=c("red", "green3", "blue")[unclass(iris$Species)])
```

## fit an LDA model to the training set data

```python
%%R

#  lda is in the MASS package

ldaModel <- lda( Species ~ Petal.Length + Petal.Width + Sepal.Length + Sepal.Width,
                 data = iris, subset = training_set )

# Equivalent abbreviation:

ldaModel <- lda( Species ~ .,  data = iris, subset = training_set )


print(ldaModel)
```

## Evaluate how well the LDA model predicts classes for the remaining data (i.e., the testing_set)

```python
%%R

# Make class predictions on the test dataset using our LDA model
#  and then check the predicted class results against the true class values.

predictions <- predict( ldaModel, iris[ testing_set ,] )

print(predictions$class)

incorrect.prediction  <-  (predictions$class != iris$Species[testing_set])

print(testing_set[ incorrect.prediction ])          # ids  for which prediction was incorrect

print(iris[ testing_set[incorrect.prediction], ])   # data for which prediction was incorrect

confusion_matrix <- table( iris$Species[testing_set], predictions$class )
print(confusion_matrix)

m <- length(testing_set)
print(m)

accuracy <- (m - sum(incorrect.prediction))/m
print(accuracy)
```

## Never mix the training set with the testing set !!

```python
%%R

#  out of curiousity, let us see how well the model works on the training set.
#  (It is a terrible idea to do this in general)

predictions <- predict( ldaModel, iris[ training_set, ] )
print(predictions$class)

incorrect.prediction  <-  (predictions$class != iris$Species[training_set] )

print(iris_ids[ incorrect.prediction ]) # ids  for which prediction was incorrect
print(iris[ incorrect.prediction , ])   # data for which prediction was incorrect

accuracy <- (n - sum(incorrect.prediction))/n
print(accuracy)

confusion_matrix <- table( iris$Species[training_set], predictions$class )
print(confusion_matrix)
```

## Training another kind of model:  the SVM

```python
%%R

# Now train an SVM model, and evaluate its accuracy

library(e1071)

svmModel <- svm( Species ~ ., data=iris, subset = training_set )
print(summary(svmModel))

svm.predictions <- predict( svmModel, iris[ testing_set ,] )

svm.incorrect.prediction  <-  (svm.predictions != iris$Species[testing_set])
print(testing_set[svm.incorrect.prediction])

svm.confusion.matrix <- table( iris$Species[testing_set], svm.predictions )
print(svm.confusion.matrix)

svm.accuracy <- (m - length(training_set[svm.incorrect.prediction])) / m
print(svm.accuracy)
```

# Problem 3.  Do problem 10 in [ISL] Chapter 4, for the Weekly dataset.

```python
%%R
#(a)
data(Weekly)

print(summary(Weekly))

print(dim(Weekly))
pairs(~Year+Lag1+Lag2+Lag3+Lag4+Lag5+Volume+Today+Direction,data = Weekly, main="Scatterplot Matrix")

```

Seems that "Year" has some exponential relationship with volume.

```python
%%R
#(b)
Weekly$Direction = factor(Weekly$Direction, levels = c("Down","Up"))
logit = glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume, data = Weekly, family = "binomial")
summary(logit)

```

Lag3 is the most significant feature but it is still not very effective.

```python
%%R
#(c)
testd = predict(logit, Weekly, type="response")
glm.pred =rep("Down",length(Weekly$Direction)) 
glm.pred[testd>.5]="Up"
confMat = table(glm.pred,Weekly$Direction)
print(confMat)
correct_fr = (confMat[1,1]+confMat[2,2])/length(testd)
cat(sprintf("The correct rate is: %f", correct_fr))
```

Answer: overall correct fraction is (54+557)/1089 = 0.56. The result is terrible. Logistic regression has too many false positive and only works better than random guess. Also, here is the confusion matrix for the training set, which underestimates the read error rate.

```python
%%R
#(d)
train = (Weekly$Year>=1990&Weekly$Year<=2008)
Weekly.2009 = Weekly[!train,]
Direction.2009 = Weekly$Direction[!train]
logit = glm(Direction~Lag2, data =Weekly, family = "binomial",subset = train)
testd = predict(logit, Weekly.2009,type="response")
glm.pred =rep("Down",length(testd)) 
glm.pred[testd>.5]="Up"
confMat = table(glm.pred,Direction.2009)
print(confMat)
correct_fr = (confMat[1,1]+confMat[2,2])/length(testd)
cat(sprintf("The correct rate is: %f", correct_fr))
```

Answer: overall correct fraction is around 0.625, which is slightly better than (c)

```python
%%R
#(e)
library(MASS)
ldaModel <- lda( Direction ~ Lag2,
                 data = Weekly, subset=train )
testd = predict( ldaModel, Weekly.2009 )
names(testd)

confMat = table(testd$class,Direction.2009)
print(confMat)
correct_fr = (confMat[1,1]+confMat[2,2])/sum(confMat)
cat(sprintf("The correct rate is: %f", correct_fr))

```

```python
%%R
#(f)

qdaModel <- qda( Direction ~ Lag2,
                 data = Weekly, subset = train )
testd = predict( qdaModel, Weekly.2009 )


confMat = table(testd$class,Direction.2009)
print(confMat)
correct_fr = (confMat[1,1]+confMat[2,2])/sum(confMat)
cat(sprintf("The correct rate is: %f", correct_fr))
```

```python
%%R
#(g)

library(class)
train.Direction = Weekly$Direction[train];
train.X = cbind(Weekly$Lag2)[train,]
test.X=cbind(Weekly$Lag2)[!train,]
print(length(train.X))
print(length(test.X))
set.seed(1)
knn.pred=knn(data.frame(train.X),data.frame(test.X),train.Direction, k=1)
confMat=table(knn.pred,Direction.2009)
print(confMat)
correct_fr = (confMat[1,1]+confMat[2,2])/sum(confMat)
cat(sprintf("The correct rate is: %f", correct_fr))
```

It's the worst result, which is equal to random guess

Anwser for(h): The LDA and Logistic regression work the best.


```python
%%R
#(i)
knn.pred=knn(data.frame(train.X),data.frame(test.X),train.Direction, k=2)
confMat=table(knn.pred,Direction.2009)
print(confMat)
correct_fr = (confMat[1,1]+confMat[2,2])/sum(confMat)
cat(sprintf("The correct rate is: %f", correct_fr))

```

use knn with k=2 is much better!

# Problem 4.  Develop LDA, QDA, SVM, and logistic regression models for the SPAM dataset (in the ElemStatLearn library)

```python
%%R

data(spam)

#print(summary(spam))

contrasts(spam$spam)
n=dim(spam)[1]
spam_ids = (1:n)

train = sample(spam_ids, 4000)
test = spam_ids[-train]



# LDA
ldam = lda(spam~.,data=spam, subset=train)
pred.lda = predict(ldam, spam[test,])
confMat = table(pred.lda$class,spam[test,]$spam)
cat("LDA for spam classification")
print(confMat)
correct_fr = (confMat[1,1]+confMat[2,2])/sum(confMat)
cat(sprintf("The correct rate is: %f\n", correct_fr))

# QDA
qdam = qda(spam~.,data=spam, subset=train)
pred.qda = predict(qdam, spam[test,])
confMat = table(pred.qda$class,spam[test,]$spam)
cat("QDA for spam classification")
print(confMat)
correct_fr = (confMat[1,1]+confMat[2,2])/sum(confMat)
cat(sprintf("The correct rate is: %f\n", correct_fr))



# ........ (note section 4.4.5 of [ESL] compares LDA and logistic regression)



```
